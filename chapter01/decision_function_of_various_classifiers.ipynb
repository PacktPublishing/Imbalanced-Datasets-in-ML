```python
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
import pandas as pd
from collections import Counter

X, y = make_classification(n_samples=10000, n_features=2,
                          n_redundant=0, n_classes=2, flip_y=0,
                          n_clusters_per_class=2, class_sep=0.89,
                          weights=[0.99], random_state=81)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.neural_network import MLPClassifier
from xgboost import XGBClassifier
from mlxtend.plotting import plot_decision_regions
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

# Define classifiers and their labels
classifiers = {
    'Logistic Regression': LogisticRegression(),
    'Decision Tree': DecisionTreeClassifier(),
    'Random Forest': RandomForestClassifier(),
    'SVM': SVC(gamma='auto'),
    'XGBoost': XGBClassifier(),
    'Neural Network': MLPClassifier(),
}

# Initialize plot
fig = plt.figure(figsize=(16, 14))
gs = gridspec.GridSpec(3, 2)

# Train classifiers and plot their decision regions
for idx, (label, classifier) in enumerate(classifiers.items()):
    classifier.fit(X, y)
    ax = plt.subplot(gs[idx // 2, idx % 2])
    plot_decision_regions(X=X, y=y, clf=classifier, legend=0)
    plt.title(label)

plt.show()
```
